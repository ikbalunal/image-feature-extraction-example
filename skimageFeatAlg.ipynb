```python
from skimage.color import rgb2gray
from skimage.feature import CENSURE

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import accuracy_score
from skimage.transform import resize
from skimage.io import imread
```

```python

import matplotlib.image as mpimg  

img="/content/drive/MyDrive/Colab Notebooks/generalProject/DatasetForImageClassification/cars/cars_1.jpg"
img_array= imread(img)
img_resized=resize(img_array,(150,150,3))
img_resized = rgb2gray(img_resized)

img_resized = img_resized.astype('double')

detector = CENSURE()

detector.detect(img_resized)

print(type(img_array), "img")
print(type(img_resized), "img_resized type")
```

```python
from skimage.feature import corner_harris, corner_subpix, corner_peaks

coords = corner_peaks(corner_harris(img_resized), min_distance=5)
coords_subpix = corner_subpix(img_resized, coords, window_size=25)

plt.subplot(121)
plt.title('CENSURE ozellik cikartma')

# cmap= 'Set3'
plt.imshow(img_resized, cmap='gist_earth')

plt.scatter(detector.keypoints[:, 1], detector.keypoints[:, 0],2 ** detector.scales, facecolors='none', edgecolors='r')
              

plt.show()
```
